```python
import pandas as pd

df = pd.read_pickle("/home/robin/Code_repo/nlp_pipeline/outputs/nlp/stanza/CAE0001_part_1_1.pkl")
```

```python
def build_nlp_dict_from_df(df,speaker):
    df_speaker = df[df['speaker']==speaker]

    text = '\n'.join(df_speaker['text']).replace('*','')
    token = df_speaker['token'].sum()
    lemma = df_speaker['lemma'].sum()
    pos = df_speaker['pos'].sum()
    morph = df_speaker['morph'].sum()
    try :
       doc = df_speaker['doc'].tolist()
    except :
        doc = []
    result = {
      'text': text,
      'token': token,
      'lemma':lemma,
      'pos':pos,
      "morph" : morph,
      "doc":doc
    }
    result['speaker']  = speaker
    
    return result
```

```python
build_nlp_dict_from_df(df, speaker='enqueteur')
```
